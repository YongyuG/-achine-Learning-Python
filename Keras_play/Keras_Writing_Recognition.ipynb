```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense,Dropout,Activation
from keras.layers import Conv2D,MaxPooling2D,Flatten
from keras.optimizers import SGD,Adam
from keras.utils import np_utils
from keras.datasets import mnist
```

```python
def load_data():
    (x_train,y_train),(x_test,y_test)=mnist.load_data()
    number=10000
    x_train=x_train[0:number]
    y_train=y_train[0:number]
    
    x_train=x_train.reshape(number,28*28)
    x_test=x_test.reshape(x_test.shape[0],28*28)
    x_train=x_train.astype('float32')
    x_test=x_test.astype('float32')
    #convert class vectors to binary class matrices
    
    y_train=np_utils.to_categorical(y_train,10)
    y_test=np_utils.to_categorical(y_test,10)
    
    x_train=x_train
    x_test=x_test
    #normalize,原本数值是0-255，除于255使之normalize　变成０－１之间
    x_train=x_train/255
    x_test=x_test/255
    
    #给testing data加上一个noise
    #x_test = np.random.normal(x_test)

    return(x_train,y_train),(x_test,y_test)
```

```python
(x_train,y_train),(x_test,y_test)=load_data()
x_train.shape
#x_train 是２维的向量，总共有10000笔training data,每笔有７８４维的vector
```

```python
model=Sequential()
model.add(Dense(input_dim=28*28,units=689,activation='relu'))
#model.add(Dropout(0.7))
model.add(Dense(units=689,activation='relu'))
#model.add(Dropout(0.7))
model.add(Dense(units=689,activation='relu'))
#model.add(Dropout(0.7))

#for i in range(10):
#    model.add(Dense(units=689,activation='relu'))

model.add(Dense(units=10,activation='softmax'))

#Adam 收敛converge的素的会比较快
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
#model.compile(loss='categorical_crossentropy',optimizer=SGD(lr=0.1),metrics=['accuracy'])

model.fit(x_train,y_train,batch_size=100,epochs=20)

result=model.evaluate(x_train,y_train,batch_size=10000)
print('Train Acc: ',result[1])

result=model.evaluate(x_test,y_test,batch_size=10000)
print('Test Acc: ',result[1])
```
